```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation, rc
rc('animation', html='html5')
```

```
fig = plt.figure()
ax = plt.axes(xlim=(-10.0*np.pi, 10.0*np.pi), ylim=(-1, 1))

ax.set_xticks([-10.0*np.pi,-3.0*np.pi, 0, 3.0*np.pi, 10.0*np.pi])
ax.set_xticklabels(['$-10\pi$','$-3\pi$','$0$','$3\pi$','$10\pi$'])
ax.set_yticks([-1, -0.6, 0, 0.6, 1])
ax.set_xlabel('$x$')
ax.set_ylabel('$u(x,t)$')

plt.close()
```

```
line, = ax.plot([], [], lw=2)
def init():
    line.set_data([], [])
    return (line,)

c = 1
def animate(t):
    x = np.linspace(-10.0*np.pi, 10.0*np.pi, 50)
    u = np.sin(2.0*np.pi*x) * np.exp(-0.1*t)
    line.set_data(x, u)
    return (line,)

tt = list(np.linspace(0, 100, 100))
anim = animation.FuncAnimation(fig, animate, init_func=init, frames=tt, interval=20, blit=True)
anim
```
